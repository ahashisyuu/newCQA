```python
import pickle as pkl
import os

savepath = '../data'

with open(os.path.join(savepath, 'dataset.pkl'), 'rb') as fr:
    all_samples_concate = pkl.load(fr)

keys_lists = ['15dev.xml', '15test.xml', '15train.xml', 
              '16dev.xml', '16test.xml', '16train1.xml', '16train2.xml', '17test.xml']
```

```python
all_samples_concate['15dev.xml'].head(2)
print(all_samples_concate['15dev.xml'].columns)
```

```python
import pandas as pd

all_triplets_samples = {}

for key in keys_lists:
    print('\n\n', key)
    samples = all_samples_concate[key]
    grouped = samples.groupby('q_id')
    # q_id, a_p_id, a_n_id, q, q_len, a_p, a_p_len, a_n, a_n_len
    new_samples = []
    for q_id, group in grouped:
        ans_grouped = group.groupby('Rrel_index')
        ans_grouped = [a for a in ans_grouped]
#         ans_pos = ans_grouped.get_group(0)
#         ans_neg = ans_grouped.get_group(1)
        ans_grouped_ = {a[0]: a[1] for a in ans_grouped}
        assert len(ans_grouped) < 3
        if len(ans_grouped) == 2:
            ans_pos = ans_grouped_[0]
            ans_neg = ans_grouped_[1]
            q_samples = []
            for i in range(len(ans_pos)):
                pos_line = ans_pos.iloc[i]
                a_p_id = pos_line['c_id']
                q_cate = pos_line['cate_index']
                q = pos_line['qTEXT_lemma_index']
                q_len = pos_line['qTEXT_len']
                a_p = pos_line['cTEXT_lemma_index']
                a_p_len = pos_line['cTEXT_len']
                for j in range(len(ans_neg)):
                    neg_line = ans_neg.iloc[j]
                    a_n_id = neg_line['c_id']
                    a_n = neg_line['cTEXT_lemma_index']
                    a_n_len = neg_line['cTEXT_len']
                    q_samples.append([q_id, a_p_id, a_n_id, q_cate, q, q_len, a_p, a_p_len, a_n, a_n_len])
            new_samples += q_samples
        else:
            print(q_id)
    
    all_triplets_samples[key] = pd.DataFrame(data=new_samples, columns=['q_id', 'c_id', 'c_id', 'q_cate',
                                         'qTEXT_index', 'q_len', 'c_pos_index', 'c_pos_len', 'c_neg_index', 'c_neg_len'])
            
        
        
```
